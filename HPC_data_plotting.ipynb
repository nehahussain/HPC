```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= pd.read_excel("https://github.com/nehahussain/HPC/raw/main/graph.xlsx",sheet_name="Sheet1")
```

#Cuda Graphs

```python
df.head()
```

```python
optimizations = df['Optimization'].unique()

plt.figure(figsize=(8, 4))

for opt in optimizations:
    subset = df[df['Optimization'] == opt]
    plt.plot(subset['Threads per Block'], subset['Time'], marker='o', label=f'Optimization {opt}')

plt.xscale('log')
plt.xlabel('Threads per Block (log scale)')
plt.ylabel('Time')
plt.title('Time vs Threads per Block for Different Optimizations')
plt.xticks(np.sort(df['Threads per Block'].unique()))  # Modify x-axis labels
plt.gca().get_xaxis().set_major_formatter(plt.ScalarFormatter())  # Format x-axis labels
plt.legend()
plt.grid(True, which="both", ls="-")  # Grid for both major and minor ticks
plt.show()

plt.figure(figsize=(10, 5))

for opt in optimizations:
    subset = df[df['Optimization'] == opt]
    plt.plot(subset['Threads per Block'], subset['Speedup'], marker='o', label=f'Optimization {opt}')

plt.xscale('log')
plt.xlabel('Threads per Block (log scale)')
plt.ylabel('Speedup')
plt.title('Speedup vs Threads per Block for Different Optimizations')
plt.xticks(np.sort(df['Threads per Block'].unique()))  # Modify x-axis labels
plt.gca().get_xaxis().set_major_formatter(plt.ScalarFormatter())  # Format x-axis labels
plt.legend()
plt.grid(True, which="both", ls="-")  # Grid for both major and minor ticks
plt.show()

plt.figure(figsize=(10, 5))

for opt in optimizations:
    subset = df[df['Optimization'] == opt]
    plt.plot(subset['Threads per Block'], subset['Efficiency'], marker='o', label=f'Optimization {opt}')

plt.xscale('log')
plt.xlabel('Threads per Block')
plt.ylabel('Efficiency (%)')
plt.title('Efficiency vs Threads per Block for Different Optimizations')
plt.xticks(np.sort(df['Threads per Block'].unique()))  # Modify x-axis labels
plt.gca().get_xaxis().set_major_formatter(plt.ScalarFormatter())  # Format x-axis labels
plt.legend()
plt.grid(True, which="both", ls="-")  # Grid for both major and minor ticks
plt.show()
```
